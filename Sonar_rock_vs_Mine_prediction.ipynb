Importing dependencies



```python
import numpy as np        #mathematical and logical operations can be performed
import pandas as pd         #used for manipulation and data analysis/loading data into data frame
from sklearn.model_selection import train_test_split            #split data into train and test
from sklearn.linear_model import LogisticRegression         #name of the model that we are using
from sklearn.metrics import accuracy_score            #find accuracy of the model

```

Data collection and pre-processing

```python
#loading the sonar data into pandas data Frame
sonar_data=pd.read_csv("/content/Copy of sonar data (1).csv",header=None)
```

```python
sonar_data.groupby(60).mean()
```

```python
#separating  data and labels
X=sonar_data.drop(columns=60,axis=1)      #dropping the 60th columns
Y=sonar_data[60]      #storing the 60th column
```

```python
Y
```

Training and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training ----> Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
#training the logistic regression with trained data
log_model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction=log_model.predict(X_train)         #answer that we are predicting
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)       #y_tain=actual answer, we are trying to compare the scores b/w them
```

```python
print("accuracy on training data",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=log_model.predict(X_test)         #answer that we are predicting
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)  
```

```python
print("accuracy on training data",testing_data_accuracy)
```
